```python
df= pd.read_csv("heights.csv")
df.head(3)
```

```python
mean = round(df.height.mean(), 2)
std = round(df.height.std(), 2)
mean, std
```

```python
# plot a histogram for detectiong outliers
sns.histplot(df.height,kde=True)
plt.axvline(df.height.mean(), color='black', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() + std, color='red', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() - std, color='red', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() + 2*std, color='green', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() - 2*std, color='green', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() + 3*std, color='orange', linestyle='--', linewidth=1)
plt.axvline(df.height.mean() - 3*std, color='orange', linestyle='--', linewidth=1)
plt.show()
```

```python
lower = mean - 3*std
upper = mean + 3*std
lower, upper
```

```python
df[(df.height < lower) | (df.height > upper)]
```

```python
df_no_outliers = df[(df.height >= lower) & (df.height <= upper)]
df_no_outliers.shape
```

```python
sns.histplot(df_no_outliers.height,kde=True)
plt.axvline(df_no_outliers.height.mean(), color='black', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() + std, color='red', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() - std, color='red', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() + 2*std, color='green', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() - 2*std, color='green', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() + 3*std, color='orange', linestyle='--', linewidth=1)
plt.axvline(df_no_outliers.height.mean() - 3*std, color='orange', linestyle='--', linewidth=1)
plt.show()
```

```python
df['z_score'] = (df.height - df.height.mean()) / df.height.std()
df.head(10)
```

```python
df[(df.z_score < -3) | (df.z_score > 3)]
```
