```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bill_survey.csv')
df.head()
```

```python
df.City.nunique()
```

```python
sns.histplot(df.Bill,kde=True)
```

```python
sns.histplot(df[df.City=="Mumbai"].Bill, kde=True)
```

```python
sample_mean=[]
for city , subdf in df.groupby('City'):
    sample_mean.append(subdf.Bill.sample(100, replace=True).mean())

sns.histplot(sample_mean, kde=True)
    
```
